```python
fig, axes = plt.subplots(nrows=3, ncols=3, figsize=(20,20)) 
plt.subplots_adjust(hspace=0.35)

corr_to_gdp = pd.Series().astype(float) 
for col in df.columns.values[2:]:
    if((col != "GDP ($ per capita)") & (col != "Climate")):
        corr_to_gdp[col] = df["GDP ($ per capita)"].corr(df[col])

abs_corr_to_gdp = corr_to_gdp.abs().sort_values(ascending=False)

corr_to_gdp = corr_to_gdp.loc[abs_corr_to_gdp.index]

for i in range(3): 
    for j in range(3):
        sns.scatterplot(x= corr_to_gdp.index.values[i*3+j], y="GDP ($ per capita)",
                    data=df, ax=axes[i,j], marker="o")
        title= "Correlation : " + str(round(corr_to_gdp[i*3+j],2))
        axes[i,j].set_title(title)
```

```python
import plotly.graph_objs as go
from plotly.offline import init_notebook_mode, iplot
```

```python
data= dict(type="choropleth",
           locations=df.Country,
           locationmode = "country names", z= df["GDP ($ per capita)"],
           text = df.Country, colorbar={"title" : "GDP ($ per capita)"},
           colorscale="magma", reversescale= True)

layout = dict(title="GDP ($ per capita)",
              geo = dict(showframe=False , projection = {"type":"natural earth"}) )

choromap = go.Figure(data= [data], layout= layout)
iplot(choromap, validate=False)
```
